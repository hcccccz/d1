### Sorting and subsetting

#### Sorting

```python
sales.sort_values("weekly_sales")
```

```python
sales.sort_values("weekly_sales", ascending=False)
```

```python
sales.sort_values(["weekly_sales", "temperature_c"], ascending=False)
```

```python
sales.sort_values(["weekly_sales", "temperature_c"], ascending=[True,False])
```

#### Subsetting column

```python
sales["weekly_sales"]
```

```python
sales[["weekly_sales","temperature_c"]]
```

#### Subsetting rows

```python
sales["weekly_sales"]>50
```

```python
sales[sales["weekly_sales"]>50]
```

multiple conditions

```python
sales_con = sales["weekly_sales"]>50
type_con = sales["type"] == "A"

sales[sales_con & type_con]
```

```python
sales[(sales["weekly_sales"]>50& type_con) & (sales["type"] == "A")]
```

```python
sales.info()
```

Filter multiple values of categorical variable

```python
is_type_a_or_b = sales["type"].isin(["A","B"])
sales[is_type_a_or_b]
```
